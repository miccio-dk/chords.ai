```python
import pickle
import os
```

```python
directory = "/home/janek/Desktop/thesis/publication/music/chords/"
all_songs = []  
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".pkl"): 
        print(os.path.join(directory, filename))
        with open(os.path.join(directory, filename), "rb") as f:
            chords = pickle.load(f)
            all_songs.append(chords)
    else:
        continue
```

```python
all_songs[0]
```

```python
def concat_chords(song):
    return ["_".join(chord) for chord in song]
```

```python
songs_final = [concat_chords(song) for song in all_songs]
```

```python
with open("songs_final.pkl" , "wb") as f:
    pickle.dump(songs_final,f)  
```
